```python
import networkx as nx
from tqdm import tqdm
import json
import numpy as np
```

```python
base_path = '/content/drive/My Drive/GraphRL_v2/Environments/'
```

```python
N = 50 # number of nodes
p = 0.2 # probability of edge creation (Erdos-Renyi)
m = 4 # number of edges from new node (Barabasi-Albert)
```

```python
# large

synthetic_ER = {}
synthetic_BA = {}

num_train_networks = 1000

train_networks_ER = {}
train_networks_BA = {}

for idx in tqdm(range(num_train_networks)):
  G_ER = nx.erdos_renyi_graph(n = N, p = p)
  train_networks_ER[idx] = nx.node_link_data(G_ER)
  G_BA = nx.barabasi_albert_graph(n = N, m = m)
  train_networks_BA[idx] = nx.node_link_data(G_BA)

synthetic_ER['train'] = train_networks_ER
synthetic_BA['train'] = train_networks_BA

num_val_networks = 100

val_networks_ER = {}
val_networks_BA = {}

for idx in tqdm(range(num_val_networks)):
  G_ER = nx.erdos_renyi_graph(n = N, p = p)
  val_networks_ER[idx] = nx.node_link_data(G_ER)
  G_BA = nx.barabasi_albert_graph(n = N, m = m)
  val_networks_BA[idx] = nx.node_link_data(G_BA)

synthetic_ER['val'] = val_networks_ER
synthetic_BA['val'] = val_networks_BA

num_test_networks = 100

test_networks_ER = {}
test_networks_BA = {}

for idx in tqdm(range(num_test_networks)):
  G_ER = nx.erdos_renyi_graph(n = N, p = p)
  test_networks_ER[idx] = nx.node_link_data(G_ER)
  G_BA = nx.barabasi_albert_graph(n = N, m = m)
  test_networks_BA[idx] = nx.node_link_data(G_BA)

synthetic_ER['test'] = test_networks_ER
synthetic_BA['test'] = test_networks_BA

save_filename = os.path.join(base_path, 'synthetic_ER_large.json')
with open(save_filename, 'w') as f:
  json.dump(synthetic_ER, f)

save_filename = os.path.join(base_path, 'synthetic_BA_large.json')
with open(save_filename, 'w') as f:
  json.dump(synthetic_BA, f)
```
